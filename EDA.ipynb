# Tarefa 6

A empresa quer oferecer descontos aos seus clientes fiéis. Clientes que fizerem compras totalizando mais de $1.500 são considerados fiéis e vão receber um desconto.

Nosso objetivo é criar um ciclo `while` que verifique o valor total gasto e pare quando ele for atingido. Para simular novas compras, a variável `new_purchase` gera um número entre 30 e 80 em cada ciclo. Isso representa a quantidade de dinheiro gasto em uma nova compra, e é o que precisamos adicionar ao total.

Assim que o valor alvo for atingido e o ciclo `while` for encerrado, o valor final será impresso.

```python
# import da biblioteca que gera valores aleatórios
from random import randint

total_amount_spent = 1280
target_amount = 1500

while total_amount_spent < target_amount: # enquanto o valor acumulado for inferior ao target
	new_purchase = randint(30, 80)        # gera um número aleatório entre 30 e 80
	total_amount_spent += new_purchase    # valor aleatório da nova compra é somado ao valor acumulado

print(f"Novo cliente fiél identificado. Valor de compras acumulado: R$ {total_amount_spent}.")
```

# Tarefa 8

Para calcular a receita da empresa, passaremos pelas seguintes etapas:

1. Usaremos um ciclo `for` para iterar na lista `users`.
2. Extrairemos a `spendind_list` (lista de gasto)s de cada usuário e somaremos os valores.
3. Atualizaremos o `revenue` (valor da receita) com o total gasto pelos usuários.

Isso vai fornecer a receita total da empresa, que vamos imprimir no final.

```python
users = [
    ['32415', ['mike', 'reed'], 32, ['electronics', 'sport', 'books'], [894, 213, 173]], # primeira sublista
    ['31980', ['kate', 'morgan'], 24, ['clothes', 'shoes'], [439, 390]]                  # segunda sublista
]

# Para localizar a receita/revenue da empresa, é preciso somar os valores em user[4] de cada usuário, utilizando o ciclo for.

revenue = 0

for user in users:
	spendings_list = user[-1]             # extração da lista de gastos de cada usuário
	total_spendings = sum(spendings_list) # soma dos gastos extraídos de cada usuário
	revenue += total_spendings            # atualização da receita total

print(f"Receita total da empresa: R$ {revenue}.")
```
